```python
from sklearn.preprocessing import StandardScaler  # ✅ Correct import
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load dataset
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

# Apply StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_scaled, y_train, epochs=20, batch_size=8, validation_data=(X_test_scaled, y_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define mean and covariance matrix
mu = [0, 0]  # Mean vector
cov = [[2, 0.8], [0.8, 1]]  # Covariance matrix (correlation = 0.8)

# Generate random samples
samples = np.random.multivariate_normal(mu, cov, size=1000)

# Scatter plot
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bivariate Gaussian Distribution")
plt.grid()
plt.show()
```
